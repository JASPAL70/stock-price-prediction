Predict the closing price of a stock (Apple) for the next day using historical data and linear regression.

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

from google.colab import files
uploaded = files.upload()

# Step 1: Load stock data
df = pd.read_csv('AAPL1.csv')

# Step 2: Prepare data using 'Close' price
close_prices = df['Close'].values

# Create input X (previous day's price) and output y (current day's price)
X = []
y = []
for i in range(len(close_prices) - 1):
    X.append([close_prices[i]])       # shape: (n, 1)
    y.append(close_prices[i + 1])     # shape: (n,)

X = np.array(X)
y = np.array(y)

# Step 3: Create Linear Regression model
model = LinearRegression()
model.fit(X, y)

# Step 4: Make predictions
predicted = model.predict(X)

# Step 5: Plot actual vs predicted
plt.figure(figsize=(10, 5))
plt.plot(y, label='Actual Price')
plt.plot(predicted, label='Predicted Price')
plt.title('Stock Price Prediction using scikit-learn')
plt.xlabel('Days')
plt.ylabel('Stock Close Price')
plt.legend()
plt.grid(True)
plt.show()

# Show Actual vs Predicted numerically (first 10)
print("\nFirst 10 Predictions:")
print(f"{'Day':<5} {'Actual Price':<15} {'Predicted Price':<15}")
for i in range(10):
    print(f"{i+1:<5} {y[i]:<15.2f} {predicted[i]:<15.2f}")
```
